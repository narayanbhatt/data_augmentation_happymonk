```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image, ImageOps, ImageEnhance
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
import os
```

```python
image_class=os.listdir(r'images')
```

```python
b_imgpathlist=['images/butterfly/'+ file for file in os.listdir('images/'+image_class[0])]
c_imgpathlist=['images/cat/'+ file for file in os.listdir('images/'+image_class[1])]
```

```python
b_imglist=[Image.open(img_path) for img_path in b_imgpathlist]
c_imglist=[Image.open(img_path) for img_path in c_imgpathlist]
```

```python
b_reimglist=[image.resize((256,256)) for image in b_imglist]
c_reimglist=[image.resize((256,256)) for image in c_imglist]
```

```python
def flip_img(image):
    return ImageOps.flip(image)
```

```python
def mirror_img(image):
    return ImageOps.flip(image)
```

```python
def adjust_brightness(image, factor):
    enhancer=ImageEnhance.Brightness(image)
    img=enhancer.enhance(factor)
    return img
```

```python
def rotate_img(image, theta):
    return image.rotate(theta,Image.NEAREST)
```

```python
def zoom_image(image, ratio):
    (width,height)=image.size
    top=height/2-(height/ratio)/2
    bottom=height/2+(height/ratio)/2
    left=(width/2)-(width/ratio)/2
    right=(width/2)+(width/ratio)/2
    img_temp=image.crop((left,top,right,bottom))
    img=img_temp.resize((width,height))
    return img
```

```python
augmented_imglist=[]
```

```python
for i in range(2):
    image=flip_img(c_reimglist[i])
    augmented_imglist.append(image)
```

```python
for i in range(2,4):
    image=mirror_img(c_reimglist[i])
    augmented_imglist.append(image)
```

```python
for i in range(4,7):
    image=rotate_img(c_reimglist[i], 20*i)
    augmented_imglist.append(image)
```

```python
for i in range(7,10):
    image=rotate_img(c_reimglist[i], 0.1*i)
    augmented_imglist.append(image)
```

```python
for i in range(10,12):
    image=adjust_brightness(c_reimglist[i],0.8)
    augmented_imglist.append(image)
```

```python
c_reimglist.extend(augmented_imglist)
```
